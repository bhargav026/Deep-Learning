```python
import tensorflow as tf
import keras
from keras.layers import Dense
from keras.models import Sequential
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```python
X, y = make_moons(n_samples = 250, noise = 0.05, random_state = 42)
```

```python
X.shape
```

```python
plt.scatter(X[:,0], X[:,1], c= y, s =100)
plt.show()
```

```python
model = Sequential()

model.add(Dense(10, activation= 'sigmoid', input_dim=2))
model.add(Dense(10, activation= 'sigmoid'))
model.add(Dense(10, activation= 'sigmoid'))
model.add(Dense(10, activation= 'sigmoid'))
model.add(Dense(10, activation= 'sigmoid'))
model.add(Dense(10, activation= 'sigmoid'))
model.add(Dense(10, activation= 'sigmoid'))
model.add(Dense(10, activation= 'sigmoid'))
model.add(Dense(10, activation= 'sigmoid'))
model.add(Dense(10, activation= 'sigmoid'))
model.add(Dense(10, activation= 'sigmoid'))
model.add(Dense(1, activation= 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss= 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.get_weights()
```

```python
old_weights= model.get_weights()[0]
```

```python
old_weights
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state= 42)
```

```python
model.fit(X_train, y_train , epochs = 50)
```

```python
model.get_weights()[0]
```

```python
abs(((model.get_weights()[0]-old_weights)/old_weights)*100)
```

```python
 from mlxtend.plotting import plot_decision_regions
 plot_decision_regions(X_train, y_train.astype('int'), clf = model, legend = 2)
```

# reduce model complexity

```python
model1 = Sequential()

model1.add(Dense(10, activation= 'sigmoid', input_dim=2))
model1.add(Dense(10, activation= 'sigmoid'))
model1.add(Dense(10, activation= 'sigmoid'))
model1.add(Dense(1, activation = 'sigmoid'))
```

```python
model1.compile(loss= 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
old_weights = model1.get_weights()[0]
```

```python
model1.fit(X_train, y_train, epochs = 50)
```

```python
abs(((model1.get_weights()[0]-old_weights)/old_weights)*100)
```

```python
 from mlxtend.plotting import plot_decision_regions
 plot_decision_regions(X_train, y_train.astype('int'), clf = model1, legend = 2)
```

# Weight Intialization


```python
model3 = Sequential()
model3.add(Dense(10, activation = 'relu', input_dim = 2, kernel_initializer = 'he_normal'))
model3.add(Dense(10, activation = 'relu', kernel_initializer = 'he_normal'))
model3.add(Dense(10, activation = 'relu', kernel_initializer = 'he_normal'))
model3.add(Dense(10, activation = 'relu', kernel_initializer = 'he_normal'))
model3.add(Dense(1, activation = 'sigmoid'))

model3.compile(loss= 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model3.fit(X_train, y_train, epochs = 50)
```

```python
plot_decision_regions(X_train, y_train.astype('int'), clf = model3, legend = 2)
```
